```python
print(df1.Cuisines.describe)
```

```python
x=df1.Cuisines.value_counts().sort_values(ascending=False)
x
```

```python
# top 3 cuisines
z=x.head(3)
z
```

```python
sb.barplot(x=z.index,y=z.values)
plt.yticks(range(0,1200,200))
plt.xlabel("Cuisines")
plt.show()
```

```python
y=z.index
plt.pie(z,labels=y , autopct='%.2f%%',wedgeprops={'edgecolor':'black'},startangle=90)
plt.title("Cuisines")
plt.show()
```

```python
print("The most common three cuisines:",z.index)
```
