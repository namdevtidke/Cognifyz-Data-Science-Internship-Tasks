## Task: Online Delivery
Determine the percentage of restaurants that offer online delivery.

```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv("C:\\Users\\abhis\\Downloads\\Dataset .csv")
df
```

```python
df.isnull().sum()
```

```python
del_rest=df.groupby(by='Has Online delivery')['Restaurant Name'].value_counts()
del_rest
```

```python
del_rest.sum()
```

```python
del_rest_per=(df['Has Online delivery']=='Yes').sum()
```

```python
del_rest_per
```

```python
percentage=del_rest_per/del_rest.sum()*100
percentage
```
