## Task: Restaurant Chains
Identify if there are any restaurant chains present in the dataset.

```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv("C:\\Users\\abhis\\Downloads\\Dataset .csv")
df
```

```python
df.isnull().sum()
```

```python
rest_chains = df.groupby("Restaurant Name").size().reset_index(name="Chain Count")
rest_chains = rest_chains[rest_chains["Chain Count"] > 1]
```

```python
rest_chains
```

```python
rest_chains = rest_chains.sort_values(by="Chain Count", ascending=False)
```

```python
rest_chains
```

```python
top_rest_chain=rest_chains.head(10)
top_rest_chain
```

```python
plt.figure(figsize=(10, 8))
plt.bar(top_rest_chain["Restaurant Name"], top_rest_chain["Chain Count"])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.show()
```

```python
chain_ratings = df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Average Rating")
chain_votes = df.groupby("Restaurant Name")["Votes"].sum().reset_index(name="Total Votes")
```

# 2.Analyze the ratings and popularity of different restaurant chains.

```python
chain_ratings
```

```python
chain_votes
```

```python
chain_anly=pd.merge(chain_ratings, chain_votes, on="Restaurant Name")
```

```python
chain_anly
```

```python
chain_anly_sort=chain_anly.sort_values(by="Average Rating", ascending=False)
```

```python
chain_anly_sort
```

```python
chain_anly_sort.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(chain_anly_sort.head(10)["Restaurant Name"], chain_anly_sort.head(10)["Average Rating"])
plt.xticks(rotation=60, ha='right')
plt.xlabel("Restaurant Chain")
plt.ylabel("Average Rating")
plt.title("Top 10 Restaurant Chains by Average Rating")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(chain_anly_sort.head(10)["Restaurant Name"], chain_anly_sort.head(10)["Total Votes"])
plt.xticks(rotation=60, ha='right')
plt.xlabel("Restaurant Chain")
plt.ylabel("Total Votes")
plt.title("Top 10 Restaurant Chains by Total Votes")
plt.tight_layout()
plt.show()
```
